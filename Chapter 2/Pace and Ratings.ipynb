## Pace and Ratings

### Possessions and Pace

As noted by Kubatko et al. in [SPABS](https://statsbylopez.files.wordpress.com/2016/01/jqas-2007-3-3-1070.pdf), possessions represent perhaps the most important concept in analyzing basketball. By their definition, a _possession_ is a stretch of continuous control of the ball by one team. Importantly, an offensive rebound extends a possession, rather than create a new one.

While there are several nuances that make defining posssessions non-trivial (such as free throws that terminate possesions), a simplification made in both BDS and SPABS gives:

$POSS = FGA + 0.44 \times FTA - OREB + TOV$

The 0.44 number is an estimate of the fraction of free throws that end possessions from the 2002-03 through 2005-06 NBA seasons. SPABS also discusses the concept of *plays*, which differ from possesions only in that plays do not include the *OREB* term.

With possessions defined, pace is simply the number of possessions per unit time. BDS uses *total* possessions per minute across all players on a team, which isn't something I have seen elsewhere. [NBA.com](https://stats.nba.com/help/glossary/#pace) gives the usual definition of pace, which is the number of possessions per 48 minutes for a player or team.

### Ratings

Using possessions as the basis for evaluating performance, offensive and defensive ratings for team _T_ against opponent _O_ naturally follow as the number of points scored and surrendered per possession.

$ORtg_T = \frac{PTS_T}{POSS_T}$,
$DRtg_T = \frac{PTS_O}{POSS_O}$

### Data

The [stats.nba.com](https://stats.nba.com) API contains the `teamdashboardbygeneralsplits` endpoint that has tons of team performance data by different splits. You can explore the information available in the documentation [here](https://github.com/basketballrelativity/py_ball/wiki/Team).

The following leverages the `Team` class from the `py_ball` package to use data from the `teamdashboardbygeneralsplits` endpoint to replicate the pace and ratings visualizations in BDS. The `construct_season_logs` function abstracts the use of the `py_ball` to pull season data for a given team.

```python
def construct_season_logs(team_id, measure_type='Base', league_id='00'):
    """ This function pulls 2017/2018 season data for the
    team provided

    @param team_id (str): String of a 10-digit unique
        team identifier
    @param measure_type (str): One of 'Base', 'Advanced',
        'Misc', 'Scoring', 'Opponent', 'Usage' which dictates
        the type of statistics returned. For more information,
        see the documentation: https://github.com/basketballrelativity/py_ball/wiki/Team
    @param league_id (str): '00' for the NBA, '10' for the WNBA, 
        '20' for the G-League

    Returns

        - team_df (DataFrame): DataFrame containing season-level
            data for the team provided from the 2017-18 regular
            season
    """

    per_mode = 'Totals'
    season = '2017-18'

    team_data = team.Team(headers=HEADERS,
                          endpoint='teamdashboardbygeneralsplits',
                          league_id=league_id,
                          season=season,
                          measure_type=measure_type,
                          per_mode=per_mode,
                          team_id=team_id)
    team_df = pd.DataFrame(team_data.data['OverallTeamDashboard'])

    return team_df
```

As you can see, `team_df` contains common counting and rate statistics. You can disregard the fields appended with `RANK`, as this only comes into play when multiple teams or players are returned. The defensive rating formula also requires opponent data, which can be pulled with `measure_type = 'Opponent'`:

```python
oppo_df = construct_season_logs(TEAM_TO_ID_NBA['CLE'], 'Opponent')
```

Now, we can take the equations detailed above and code them up in a function, `pace_possession_calculator`.

```python
def pace_possession_calculator(game_df, prefix=''):
    """ This function calculates the possessions and pace
    from the DataFrame provided

    @param game_df (DataFrame): Single-row DataFrame
        corresponding to a team's season
    @param prefix (str): One of '' or 'OPP_'. If a team's
        data is provided, use ''. If an opponent's data
        is provided, use 'OPP_'

    Returns:

        - poss (float): Number of possessions in a season
        - pace (float): Number of possessions per 48 minutes
    """

    poss = sum(game_df[prefix + 'FGA']) \
        + (0.44 * sum(game_df[prefix + 'FTA'])) \
        - sum(game_df[prefix + 'OREB']) + sum(game_df[prefix + 'TOV'])

    pace = (48 * poss) / sum(game_df['MIN'])
    return poss, pace
```

Let's test out the above function by calculating offensive and defensive rating for Cleveland.

```python
poss, pace = pace_possession_calculator(team_df, '')
o_rtg = 100 * sum(team_df['PTS']) / poss
print('Offensive Rating')
print(round(o_rtg, 2))
print('****************')
print('Pace')
print(round(pace, 2))
```

```python
d_poss, d_pace = pace_possession_calculator(oppo_df, 'OPP_')
d_rtg = 100 * sum(oppo_df['OPP_PTS']) / d_poss
print('Defensive Rating')
print(round(d_rtg, 2))
```

The offensive and defensive ratings appear to match those in Figure 2.1 in BDS. Notably, they do not match what's available on [basketball-reference.com](https://www.basketball-reference.com/teams/CLE/2018.html). But this is very likely due to the slightly different way in which Basketball Reference calculates possessions (check out more detail in their [glossary](https://www.basketball-reference.com/about/glossary.html)). SPABS notes that sources may disagree slightly on calculating possessions due to the inexact nature of play-by-play data.

## Visualization

The following builds a wrapper function to pull pace and ratings for all teams in the 2017-18 NBA regular season to replicate Figure 2.2 in BDS. All of the visualizations leverage the `plotnine` package. To reduce the stress on the API, a sleep function is used to space out calls.

```python
def league_ratings_and_pace():
    """ This function loops over each team
    and calculates offensive and defensive rating,
    along with pace

    Returns:

        - data_dict (dict): Dictionary storing
            ratings and pace data
    """

    data_dict = {}

    for team in TEAM_TO_ID_NBA:
        data_dict[team] = {}
        print('Pulling data for ' + team)
        team_df = construct_season_logs(TEAM_TO_ID_NBA[team], 'Base')
        time.sleep(2)
        oppo_df = construct_season_logs(TEAM_TO_ID_NBA[team], 'Opponent')

        # Calculating offensive side
        poss, pace = pace_possession_calculator(team_df, '')
        o_rtg = 100 * sum(team_df['PTS']) / poss

        # Calculating defensive side
        d_poss, d_pace = pace_possession_calculator(oppo_df, 'OPP_')
        d_rtg = 100 * sum(oppo_df['OPP_PTS']) / d_poss

        # Storing data
        data_dict[team]['Offensive Rating'] = o_rtg
        data_dict[team]['Defensive Rating'] = d_rtg
        data_dict[team]['Team Pace'] = pace
        data_dict[team]['Opponent Pace'] = d_pace
        time.sleep(2)

    return data_dict
```

```python
ratings_dict = league_ratings_and_pace()
```

```python
ratings_df = pd.DataFrame(ratings_dict).T.reset_index()
ratings_df.columns = ['Team'] + list(ratings_df)[1:]
ratings_df
```

The following plots a scatter plot of offensive versus defensive ratings for each team. You may have to run the following cell a few times to avoid a lot of overlapping labels.

```python
p = ggplot(data=ratings_df) \
    + geom_point(mapping=aes(x='Offensive Rating',
                             y='Defensive Rating'), alpha=0.1) \
    + geom_text(mapping=aes(x='Offensive Rating',
                            y='Defensive Rating',
                            label='Team'),
                position=position_jitter(width=0.5,height=0.5), size=10) \
    + xlim(100, 113) \
    + ylim(100, 113)
p
```

The above allows for some quick takeaways related to team performance:
- Phoenix had a difficult 2017-18 campaign, with the worst offensive and defensive ratings
- The best teams in the league appear clustered to the lower-right portion of the chart, with higher offensive ratings and lower defensive ratings (Toronto, Golden State, and Houston)

Because the pace of a team should roughly match that of its opponents due to the nature of possessions, plotting team versus opponent pace as done in BDS seems uninformative. Instead, let's generate a bar chart.

```python
ratings_df = ratings_df.sort_values('Team Pace')
ratings_df['Teams'] = pd.Categorical(ratings_df['Team'], categories=list(ratings_df['Team']), ordered=True)
p = ggplot(data=ratings_df) \
    + geom_bar(mapping=aes(x='Teams',
                           y='Team Pace'), position='dodge', stat='identity') \
    + coord_flip()
p
```

Phoenix again shows up as an outlier, playing at the fastest pace in the league. It's difficult to compare this chart to the previous one to identify relationships between team performance and pace.
